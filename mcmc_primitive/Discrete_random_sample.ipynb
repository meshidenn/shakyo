```python
import numpy as np
from scipy.stats import uniform, poisson
import matplotlib.pyplot as plt
```

## ポアソン分布のパラメータ

```python
lam = 4
```

## この値まで確率値を計算

```python
K = 20
```

### サンプリング数

```python
N = 10000
```

```python
rv = poisson(mu=lam)
```

## cdf

```python
t = np.arange(K)
prob = rv.cdf(t)
print(t)
print(prob)
```

```python
X = []
```

```python
for i in range(N):
    u = uniform.rvs(loc=0, scale=1, size=1)
    X.append(np.sum(prob < u))
    
print(X)
```

## ポアソン分布に従う乱数の分布を描画

```python
plt.figure(1)
nbins = np.arange(-0.5, K, 1.0)
print(nbins)
weights = np.ones_like(X) / float(len(X))
plt.hist(X, nbins, weights=weights)
plt.plot(t, rv.pmf(t), 'ro-', lw=1)
plt.xlim((0, K))

plt.show()
```

## 2項分布の例

```python
from scipy.stats import binom
```

```python
n = 20
p = 0.5
```

```python
K = 40
```

```python
M = 1000-0
```

```python
rv = binom(n=n, p=p)
```

```python
t = np.arange(K)
prob = rv.cdf(t)
```

```python
X = []
```

```python
for i in range(N):
    u = uniform.rvs(loc=0, scale=1, size=1)
    X.append(np.sum(prob < u))
```

```python
plt.figure(1)
nbins = np.arange(-0.5, K, 1.0)
weights = np.ones_like(X) / float(len(X))
plt.hist(X, nbins, weights=weights)
plt.plot(t, rv.pmf(t), 'ro-', lw=1)
plt.show()
```
