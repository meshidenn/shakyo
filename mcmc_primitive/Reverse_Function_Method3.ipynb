```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special
```

```python
nbins = 50
```

## Gamma Distribution

```python
k = 2.0
theta = 2.0
```

```python
np.random.seed()
N = 100000
U = np.random.uniform(0.0, 1.0, N)
```

### 逆変換

```python
print(scipy.special.gamma(k))
X1 = theta * scipy.special.gammaincinv(k, U)
```

### 変換したガンマ分布と理想的なpdfの描画

```python
plt.figure(1)
rv = scipy.stats.gamma(a=k, scale=theta)
plt.hist(X1, nbins, normed=True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw=2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))
```

### scipyのガンマ分布と理想的なpdf

```python
plt.figure(2)
X2 = rv.rvs(N)
plt.hist(X2, nbins, normed=True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw=2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))
plt.show()
```

```python
k = 1
theta = 2
```

## Beta Distribution

```python
alpha = 5.0
beta = 1.0
```

### 逆変換

```python
np.random.seed()
N = 100000
U = np.random.uniform(0.0, 1.0, N)
```

```python
X1 = scipy.special.betaincinv(alpha, beta, U)
```

### 変換したベータ分布と理想的なpdf

```python
plt.figure(1)
rv = scipy.stats.beta(a=alpha, b=beta)
plt.hist(X1, nbins, normed=True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x,y, 'r-', lw=2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))
```

### scipyのベータ分布と理想的なpdf

```python
plt.figure(2)
X2 = rv.rvs(N)
plt.hist(X2, nbins, normed=True)
x = np.linspace(rv.pdf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x,y, 'r-', lw=2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))

plt.show()
```
